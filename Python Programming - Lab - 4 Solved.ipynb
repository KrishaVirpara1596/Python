### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input('Enter value of n:'))
count = 0
for i in range(2,n+1):
    if n%i == 0:
        count = count + 1
if count == 1:
    print('Prime')
else:
    print('Not prime')
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input('Enter value of n:'))
sum = 0
while n!=0:
    rem = n%10
    sum = sum + rem
    n = n//10
print(sum)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
 n = int(input('Enter value of n:'))
og = n
ans = 0
while og!=0:
    rem = og%10
    ans = ans*10 + rem
    og = og//10
if ans == n:
    print('Palindrome')
else:
    print('Not')
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = int(input('Enter value of n:'))
og = n
count = 0 
sum =0
while og!=0:
    count = count+1
    og = og//10
og = n
while og!=0:
    rem = og%10
    pow = 1
    for i in range(1,count+1):
        pow =pow*rem
    sum =sum+pow
    og = og//10
if sum == n:
    print('Armstrong')
else:
    print('Not')
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input('Enter value of n:'))
for i in range(1,n+1):
    sum = 0
    for j in range(1,i):
        if i%j == 0:
            sum = sum + j
    if sum == i:
        print(i)
```
