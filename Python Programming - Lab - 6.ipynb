### 01) WAP to find sum of all the elements in a List.

```python
l = int(input('Enter value in list:'))
n = []
for i in range(0,l):
    num = int(input(f'Enter num at {i}:'))
    n.append(num)
sum = 0
for i in n:
    sum = sum + i
print(sum)

## l = (map(int,input('Enter list:').split())
```

### 14) WAP to extract elements with frequency greater than K.

```python
l = list(map(int,input('Enter list:').split()))
k = int(input('Enter value of k:'))
l1 = []
l2 = []
for i in l:
    if l.count(i) > k:
        l1.append(i) 
        if l1.count(i) == 1:
            l2.append(i)
print(l2)
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
s = int(input('Enter elements:'))
l1 = []
for i in range(0,s):
    l1.append((input('Enter list element:')))
r = [i for i in l1 if i[0] == 'b']
print(r)
```

### 17) WAP to create a list of common elements from given two lists.

```python
l1 = list(map(int,input('Enter list:').split()))
l2 = list(map(int,input('Enter list:').split()))
l3 = []
for i in l1:
    for j in l2:
        if i == j:
            l3.append(j)
print(l3)
```
