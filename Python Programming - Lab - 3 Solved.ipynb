### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input('Enter a year:'))
if year%4==0 and year%100!==0:
    print(f'{year} is leap year.')
elif year%40==0:
    print(f'{year} is leap year.')
else:
    print(f'{year} is not a leap year.')
```

### 06) WAP in python to display the name of the day according to the number given by the user.

```python
no_of_day = int (input('Enter number of day :'))
match no_of_day:
    case 1:
        print('Monday')
    case 2:
        print('Tuesday')
    case 3:
        print('Wednesday')
    case 4:
        print('Thursday')
    case 5:
        print('Friday')
    case 6:
        print('Saturday')
    case _:
        print('Invalid.')
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
a = int(input('Enter value of a:'))
b = int(input('Enter value of b:'))
choice  = int(input('Enter value of choice:'))
match choice:
    case 1:
        print(f'Addition of a and b is {a+b}')
    case 2:
        print(f'Subtraction of a and b is {a-b}')
    case 3:
        print(f'Multiplication of a and b is {a*b}')
    case 4:
        print(f'Division of a and b is {a/b}')
    case _:
        print('Invalid.')
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
eng = float(input('Enter marks of eng:')) 
maths = float(input('Enter marks of maths:')) 
sci = float(input('Enter marks of sci:')) 
comp = float(input('Enter marks of comp:')) 
ss = float(input('Enter marks of ss:')) 

per = ((eng + maths + sci + comp + ss)/500)*100
if per>=70:
    print('Distinction')
elif per >=60 and per<70:
    print('First class')
elif per >=45 and per<60:
    print('Second class')
elif per >=35 and per <45:
    print('Pass class')
elif per<35:
    print('Fail')
else:
    print('Absent')
```

### 09) WAP to find the second largest number among three user input numbers.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if (a > b and a < c) or (a > c and a < b):
    second_largest = a
elif (b > a and b < c) or (b > c and b < a):
    second_largest = b
else:
    second_largest = c

print("The second largest number is:", second_largest)
```
