### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
a = input('Enter string:')
lower = 0
upper = 0
digit = 0
special = 0
for i in a:
    if i.isupper():
        upper = upper + 1
    elif i.isdigit():
        digit = digit + 1
    elif i.islower():
        lower = lower + 1
    else:
        special = special + 1    
print(upper)
print(lower)
print(digit)
print(special)
        
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input("Enter a string: ")

result = ""
count = 1

for i in range(len(s)):
    if i + 1 < len(s) and s[i] == s[i + 1]:
        count = count + 1
    else:
        result = result + s[i] + str(count)
        count = 1   

print("Compressed string:", result)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input("Enter a string: ")

result = ""

for ch in s:
    if ch >= 'A' and ch <= 'Z':
        result = result + ch.lower()
    elif ch >= 'a' and ch <= 'z':
        result = result +  ch.upper()
    else:
        result = result + ch

print("Toggled string:", result)
```

### 11) Username Validation
#### valid username rules:
1. only alphabets and digits
2. does not start with a digit
3. length ≥ 6

- `User123` → Valid
- `123User` → Invalid
- `Us@12` → Invalid

```python
username = input("Enter username: ")

if len(username) < 6:
    print("Invalid")

elif username[0].isdigit():
    print("Invalid")

elif not username.isalnum():
    print("Invalid")

else:
    print("Valid")
```

### 13) Longest Word Finder
- `Python programming is interesting` → programming
- `I love coding` → coding
- `Data structures and algorithms` → structures

```python
s = input("Enter a sentence: ")

words = s.split() 
longest = ""

for w in words:
    if len(w) > len(longest):
        longest = w

print("Longest word:", longest)
```
